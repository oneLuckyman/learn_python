```python
import numpy as np 
import matplotlib.pyplot as plt 
plt.rcParams['font.sans-serif'] = ['FangSong']
plt.rcParams['axes.unicode_minus'] = False
```

```python
size = int(1e7)
s = np.random.uniform(-10,10,size)
sy = 1/20
k = 6
t = 0.3/np.sqrt(np.pi*2)*np.exp(-(s+2)**2/2) + 0.7/np.sqrt(np.pi*2)*np.exp(-(s-4)**2/2)
ran = np.random.uniform(0,k*sy,size)
sampling = s[ran < t]
```

```python
def f1(x):
    return 0.3/np.sqrt(np.pi*2)*np.exp(-(x+2)**2/2) + 0.7/np.sqrt(np.pi*2)*np.exp(-(x-4)**2/2)
x = np.arange(-10, 10, 0.010)
```

```python
plt.figure(figsize=(20,10))
plt.plot(x, f1(x), color='red', label='待抽样分布')
plt.plot([-10,10], [sy*k, sy*k], label='建议分布')
plt.hist(sampling, bins=200, density=True, label='抽样数据', color='green')
plt.legend(prop={'size':20})
plt.show()
```
